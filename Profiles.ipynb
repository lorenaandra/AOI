```python
def add_researcher(sheet, researcher_id, full_name, expertise_list, appreciated_list_ids):
    row_number = sheet.max_row + 1
    names = full_name.split()
    last_name = names[-1]
    first_names = ' '.join(names[:-1])
    formatted_name = f"{last_name}, {first_names}"
    sheet[f'A{row_number}'] = researcher_id
    sheet[f'B{row_number}'] = formatted_name
    expertise_string = '[' + ', '.join([f"'{exp}'" for exp in expertise_list]) + ']'
    sheet[f'C{row_number}'] = expertise_string
    sheet[f'D{row_number}'] = ', '.join(appreciated_list_ids)
```

### Sort Researchers Alphabetically by Name

```python
def get_last_name(researcher):
    return researcher['full_name'].split()[-1]

sorted_researchers = sorted(new_researchers, key=get_last_name)
```

### Save Data

```python
if os.path.exists(file_path):
    os.remove(file_path)

# headers
workbook = openpyxl.Workbook()
sheet = workbook.active
sheet['A1'] = 'Researcher ID'
sheet['B1'] = 'Full Name'
sheet['C1'] = 'Expertise'
sheet['D1'] = 'Appreciated'


for researcher in sorted_researchers:
    add_researcher(sheet, researcher['researcher_id'], researcher['full_name'], 
                   researcher['expertise_list'], researcher['appreciated_list_ids'])



# fit column width to text width
for column in sheet.columns:
    max_length = 0
    column = [cell for cell in column]
    for cell in column:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(cell.value)
        except:
            pass
    adjusted_width = (max_length + 2)
    sheet.column_dimensions[column[0].column_letter].width = adjusted_width


workbook.save(file_path)
```
